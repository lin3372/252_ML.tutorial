### Dynamic Programming

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n - 1) + fib(n - 2)
```

```python
fib(7)
```

```python
mem = {0:0, 1:1}
def fib_mem(n):
    if n not in mem:
        mem[n] = fib(n - 1) + fib(n - 2)
    return mem[n]

mem8 = {0:0, 1:1, 2:1, 3:2, 4:3, 5:5, 6:8, 7:13, 8:21}
def fib_mem8(n):
    if n not in mem8:
        mem8[n] = fib(n - 1) + fib(n - 2)
    return mem8[n]
```

```python
fib_mem(7)
```

### Time Performance

```python
n=35
```

```python
%timeit fib(n)
# We get 5.54 seconds to run with n=35
```

```python
%timeit fib_mem(n)
# 1st run - We get 500 ns to run with n=35

%timeit fib_mem(n)
# 2nd run - We get 174 ns to run with n=35
```

```python
%timeit fib_mem8(n)
# 1st run - We get 655 ns to run with n=35

%timeit fib_mem8(n)
# 2nd run - We get 206 ns to run with n=35
```
